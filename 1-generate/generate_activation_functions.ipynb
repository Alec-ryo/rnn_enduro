```python
import numpy as np  
import matplotlib.pyplot as plt  
import math  
```

```python
def sigmoid(x): return 1/(1 + np.exp(-x))
def tanh(x): return np.sinh(x)/np.cosh(x)
def relu(x): return np.maximum(0, x)
```

```python
def degrau(x):
    return 1 if x >= 0 else 0
```

```python
import matplotlib.pyplot as plt

x = [-2,-1, 0, 1, 2] 
y = [ 0, 0, 0, 1, 1]

plt.step(x, y)
plt.show()
```

```python
x = np.linspace(-2,2)  
plt.plot(x, sigmoid(x)) 
plt.ylabel('output')
plt.xlabel('input')
plt.legend(['sigmoid'], loc='upper left')
plt.show()
```

```python
x = np.linspace(-2,2)  
plt.plot(x, x) 
plt.plot(x, sigmoid(x)) 
plt.plot(x, tanh(x))
plt.plot(x, relu(x))
plt.ylabel('output')
plt.xlabel('input')
plt.legend(['linear', 'sigmoid', 'tanh', 'relu'], loc='upper left')
plt.grid()  
plt.show()
```
