```python
import numpy as np  
import matplotlib.pyplot as plt  
import math  
```

```python
def sigmoid(x): return 1/(1 + np.exp(-x))
def tanh(x): return np.sinh(x)/np.cosh(x)
def relu(x): return np.maximum(0, x)
```

```python
x = np.linspace(-2,2)  
plt.plot(x, x) 
plt.plot(x, sigmoid(x)) 
plt.plot(x, tanh(x))
plt.plot(x, relu(x))
plt.ylabel('output')
plt.xlabel('input')
plt.legend(['linear', 'sigmoid', 'tanh', 'relu'], loc='upper left')
plt.grid()  
plt.show()
```
